```python
import pandas as pd

sofa = pd.read_csv("../data/SOFA_SCORE.csv")

sofa = sofa.drop(['hadm_id', 'sofa_start', 'sofa_end'], axis=1)
sofa = sofa.groupby(['stay_id']).max()
sofa.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for i, col in enumerate(sofa.columns.values):
    plt.subplot(2, 3, i+1)
    sofa[col].value_counts().sort_index().plot(kind='bar', title=f'{col}')
    plt.xlabel('SOFA Score')
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
